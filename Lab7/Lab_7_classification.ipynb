```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from datasets import load_dataset
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn import metrics
from collections import Counter
import transformers
import numpy as np
import fasttext
import ktrain
from ktrain import text
from imblearn.over_sampling import RandomOverSampler
```

# 1. Get acquainted with the data of the Polish Cyberbullying detection dataset. Pay special attention to the distribution of the positive and negative examples in the first task as well as distribution of the classes in the second task.

```python
def get_results(y_test, y_pred):
    print("Confusion matrix: ")
    print(metrics.confusion_matrix(y_test, y_pred))
    # sklearn.metrics.ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
    # plt.show()
    print("Classification: ")
    print(metrics.classification_report(y_test, y_pred))
    print("MCC: ")
    print(metrics.matthews_corrcoef(y_test, y_pred))
```

# 2. Train the following classifiers on the training sets (for the task 1 and the task 2):

- Bayesian classifier with TF * IDF weighting.
- Fasttext text classifier
- Transformer classifier (take into account that a number of experiments should be performed for this model).


# Task01

```python
dataset = load_dataset("poleval2019_cyberbullying", "task01")
```

```python
x_train = dataset["train"]['text']
x_test = dataset["test"]['text']

vectorizer = TfidfVectorizer()
train_vector = vectorizer.fit_transform(x_train).toarray()
test_vector = vectorizer.transform(x_test).toarray()

gnb = GaussianNB()
mnb = MultinomialNB()

y_pred = gnb.fit(train_vector, dataset["train"]["label"]).predict(test_vector)
print("GaussianNB")
get_results(dataset["test"]["label"], y_pred)

y_pred = mnb.fit(train_vector, dataset["train"]["label"]).predict(test_vector) # raczej multinomial jest uzywany do tekstu
print("MultinomialNB")
get_results(dataset["test"]["label"], y_pred)
```

```python
print(Counter(dataset["train"]["label"]))
# set jest niezbalansowany
```

```python
# Przygotowanie pliku pod fasttext
x_train = dataset["train"]["text"]
y_train = dataset["train"]["label"]

with open("train_data.txt", 'wb') as f:
    f.writelines([f"__label__{label} {sentence}\n".encode() for sentence, label in zip(x_train, y_train)])
```

```python
model = fasttext.train_supervised("train_data.txt")

y_pred, _ = model.predict(x_test)
y_pred = [int(label[0].split("__label__")[1]) for label in y_pred]
y_test = dataset["test"]["label"]
get_results(dataset["test"]["label"], y_pred)
```

```python
# https://github.com/amaiya/ktrain
```

```python
x_train = dataset["train"]["text"]
y_train = dataset["train"]["label"]
x_test = dataset["test"]["text"]
y_test = dataset["test"]["label"]
```

```python
def transformer_model(model_name, x_train, y_train, x_test):
    # https://nbviewer.org/github/amaiya/ktrain/blob/develop/tutorials/tutorial-A3-hugging_face_transformers.ipynb
    t = ktrain.text.Transformer(model_name, maxlen=25, class_names=np.unique(y_train)) # 25 bo pokazuje ze niby 24 jest max
    train = t.preprocess_train(x_train, y_train)
    model = t.get_classifier()
    learner = ktrain.get_learner(model, train_data=train, batch_size=32)
    learner.fit_onecycle(5e-5, 2)
    
    predictor = ktrain.get_predictor(learner.model, preproc=t)
    y_pred = predictor.predict(x_test)
    
    return y_pred
```

### Customowe

```python
y_pred = transformer_model('dkleczek/bert-base-polish-cased-v1', x_train, y_train, x_test)
get_results(dataset["test"]["label"], y_pred)
```

```python
y_pred = transformer_model('allegro/herbert-large-cased', x_train, y_train, x_test)
get_results(dataset["test"]["label"], y_pred)
```

```python
y_pred = transformer_model('Geotrend/distilbert-base-pl-cased', x_train, y_train, x_test)
get_results(dataset["test"]["label"], y_pred)
```

```python
y_pred = transformer_model('Geotrend/bert-base-pl-cased', x_train, y_train, x_test)
get_results(dataset["test"]["label"], y_pred)
```

### Bazowe

```python
y_pred = transformer_model('xlm-roberta-base', x_train, y_train, x_test)
get_results(dataset["test"]["label"], y_pred)
```

```python
y_pred = transformer_model('distilbert-base-multilingual-cased', x_train, y_train, x_test)
get_results(dataset["test"]["label"], y_pred)
```

```python
y_pred = transformer_model('bert-base-multilingual-cased', x_train, y_train, x_test)
get_results(dataset["test"]["label"], y_pred)
```
